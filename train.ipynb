```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# We load the data in a DataFrame
train = pd.read_csv("C:/Users/train.csv")
```

```python
# We print some information about the data
print('train is of type:', type(train))
print('train has shape:', train.shape)
```

```python
#display the dataset
train
```

```python
#display the last 7 rows of the dataset
train.tail(7)
```

```python
#check for NaN values in the dataset
train.isnull().any()
```

```python
#bar chart to depict the distribution of the 'Survived' categorical variable 
sb.countplot(data = train, x = 'Survived');
```

```python
#bar chart to depict the distribution of the 'Embarked' categorical variable 
base_color = sb.color_palette()[2]
sb.countplot(data = train, y = 'Embarked', color = base_color)
```

```python
#count the number of missing values in each column
train.isna().sum()
```

```python
# pie chart for the 'Pclass' categorical variable
sorted_counts = train['Pclass'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index);
```

```python
#donut plot for the 'Age' categorical variable
sorted_counts = train['Age'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, wedgeprops = {'width' : 0.4});
plt.axis('square');
```

```python
#histogram for the 'Fare' numerical variable
plt.hist(data = train, x = 'Fare')
```
